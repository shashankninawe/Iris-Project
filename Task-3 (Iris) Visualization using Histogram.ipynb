# Create Matplot

```python
plt.hist(data['SepalLengthCm'], bins=10, edgecolor='k', color='purple', alpha=0.7)
plt.xlabel('Sepal length Cm')
plt.ylabel('Frwquency')
plt.title('Histogram of Sepal Length')
plt.show()
```

```python
plt.hist(data['SepalWidthCm'], bins=10, edgecolor='k',color='red', alpha=0.7)
plt.xlabel('Sepal length Cm')
plt.ylabel('Frwquency')
plt.title('Histogram of Sepal Width')
plt.show()
```

```python
plt.hist(data['PetalLengthCm'], bins=10, edgecolor='k', color='pink', alpha=0.7)
plt.xlabel('Petal Length Cm')
plt.ylabel('Frwquency')
plt.title('Histogram of Petal Length')
plt.show()
```

```python
plt.hist(data['PetalWidthCm'], bins=10, edgecolor='k',color='orange', alpha=0.7)
plt.xlabel('Petal Width Cm')
plt.ylabel('Frwquency')
plt.title('Histogram of Petal Width')
plt.show()
```

```python
species_count = data['Species'].value_counts()
plt.bar(species_count.index, species_count.values,color='green', alpha=0.7)
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.title('Distribution of Species')
plt.show()
```
